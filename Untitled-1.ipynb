```python
import pandas as pd
```

```python
def read_data(file_path):
    df = pd.read_csv(file_path)

    # drop non-relevant columns
    df.drop(columns=['Reason','index'],inplace=True)

    # # fill missing values
    df['day'].fillna(df['day'].mode()[0],inplace=True)
    df['month'].fillna(df['month'].mode()[0],inplace=True)
    df['year'].fillna(df['year'].mode()[0],inplace=True)
    df['Temperature'].fillna(df['Temperature'].mode()[0],inplace=True)
    df[' RH'].fillna(df[' RH'].mode()[0],inplace=True)
    df[' Ws'].fillna(df[' Ws'].mode()[0],inplace=True)
    df['Rain '].fillna(df['Rain '].mode()[0],inplace=True)
    df['FFMC'].fillna(df['FFMC'].mode()[0],inplace=True)
    df['DMC'].fillna(df['DMC'].mode()[0],inplace=True)
    df['DC'].fillna(df['DC'].mode()[0],inplace=True)
    df['ISI'].fillna(df['ISI'].mode()[0],inplace=True)
    df['BUI'].fillna(df['BUI'].mode()[0],inplace=True)
    df['FWI'].fillna(df['FWI'].mode()[0],inplace=True)
    df['Classes  '] = df['Classes  '].str.strip()
    df['Classes  '].fillna(df['Classes  '].mode()[0],inplace=True)

    #replase value
    df['Classes  '].replace('not  fire', 'not fire', inplace=True)
    df['Classes  '].replace('Fire', 'fire', inplace=True)
    df['Classes  '].replace('TRUE', 'fire', inplace=True)
    df['Classes  '].replace('Classes', 'fire', inplace=True)

    # change type to int
    df['day'] = pd.to_numeric(df['day'], errors='coerce')
    df['day'] = df['day'].fillna(df['day'].mode()[0]).astype(int)
    df['month'] = pd.to_numeric(df['month'], errors='coerce')
    df['month'] = df['month'].fillna(df['month'].mode()[0]).astype(int)
    df['year'] = pd.to_numeric(df['year'], errors='coerce')
    df['year'] = df['year'].fillna(df['year'].mode()[0]).astype(int)
    df[' RH'] = pd.to_numeric(df[' RH'], errors='coerce')
    df[' RH'] = df[' RH'].fillna(df[' RH'].mode()[0]).astype(int)
    df[' Ws'] = pd.to_numeric(df[' Ws'], errors='coerce')
    df[' Ws'] = df[' Ws'].fillna(df[' Ws'].mode()[0]).astype(int)
    # change type to float
    df['Temperature'] = pd.to_numeric(df['Temperature'], errors='coerce')
    df['Temperature'] = df['Temperature'].fillna(df['Temperature'].mode()[0]).astype(float)
    df['Rain '] = pd.to_numeric(df['Rain '], errors='coerce')
    df['Rain '] = df['Rain '].fillna(df['Rain '].mode()[0]).astype(float)
    df['FFMC'] = pd.to_numeric(df['FFMC'], errors='coerce') 
    df['FFMC'] = df['FFMC'].fillna(df['FFMC'].mode()[0]).astype(float)
    df['DC'] = pd.to_numeric(df['DC'], errors='coerce')
    df['DC'] = df['DC'].fillna(df['DC'].mode()[0]).astype(float)
    df['ISI'] = pd.to_numeric(df['ISI'], errors='coerce')
    df['ISI'] = df['ISI'].fillna(df['ISI'].mode()[0]).astype(float)
    df['BUI'] = pd.to_numeric(df['BUI'], errors='coerce')
    df['BUI'] = df['BUI'].fillna(df['BUI'].mode()[0]).astype(float)
    df['FWI'] = pd.to_numeric(df['FWI'], errors='coerce')
    df['FWI'] = df['FWI'].fillna(df['FWI'].mode()[0]).astype(float)
    df['DMC'] = pd.to_numeric(df['DMC'], errors='coerce')
    df['DMC'] = df['DMC'].fillna(df['DMC'].mode()[0]).astype(float)

    # change to date
    df['date'] = pd.to_datetime(df[['year', 'month', 'day']])
  

    return df
df = read_data(r'C:\Users\anwer\OneDrive\Desktop\NEW DATA\forest_fire.csv')
```
